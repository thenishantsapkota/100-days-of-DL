```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./placement.csv")
df.shape
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(df['cgpa'], df["resume_score"], c=df["placed"], cmap="winter", s=100)
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
def perceptron(X, y):
    w1 = w2 = b = 1
    lr = 0.1

    for j in range(100_000):
        for i in range(X.shape[0]):
            # check condition
            z = w1 * X[i][0] + w2 * X[i][1] + b

            if z * y[i] < 0:
                w1 = w1 + lr * y[i] * X[i][0]
                w2 = w2 + lr * y[i] * X[i][1]
                b = b + lr * y[i]

    return w1, w2, b
```

```python
w1, w2, b = perceptron(np.array(X), np.array(y))
```

```python
m = -(w1/w2)
c = -(b/w2)
print(m,c)
```

```python
x_min = df['cgpa'].min() - 1
x_max = df['cgpa'].max() + 1
x_input = np.linspace(x_min, x_max, 500)
y_input = m*x_input + b
```

```python
# Plotting the initial data and the line
plt.figure(figsize=(20,10))
plt.plot(x_input, y_input, color="red", linewidth=3)
plt.scatter(df["cgpa"], df["resume_score"], c=df["placed"], cmap="winter", s=50)
```
